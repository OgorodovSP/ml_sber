#### GA Sessions DataSet


```python
# load 'ga_sessions' dataset
%time
ga_sessions = pd.read_csv('../data/ga_sessions.csv')
print(ga_sessions.shape)
ga_sessions.sample(5)
```

#### GA Hits DataSet

```python
# load 'ga_hits' dataset
%time
ga_hits = pd.read_csv('../data/ga_hits.csv')
print(ga_hits.shape)
ga_hits.head(5)
```

```python
#checking duplicates
ga_hits.duplicated().sum()
```

```python
#see value_counts in column 'event_action'
ga_hits.event_action.value_counts()
```

```python
#creating target action list
target_action_list = ['sub_car_claim_click', 'sub_car_claim_submit_click',
'sub_open_dialog_click', 'sub_custom_question_submit_click',
'sub_call_number_click', 'sub_callback_submit_click', 'sub_submit_success',
'sub_car_request_submit_click']
```

```python
#filtering ga_hits.csv by target_action_list
df_hits_with_target_action = ga_hits.loc[ga_hits['event_action'].isin(target_action_list)]
df_hits_with_target_action.head(5)
```

```python
#see result. Only 104 908 row from 15 726 470 have target actions
df_hits_with_target_action.shape
```

```python
#see list of columns df. Think what columns need drops
list(df_hits_with_target_action.columns)
```

```python
# Creating list to drop. drop them
cols_to_drop = ['hit_date','hit_time','hit_number','hit_type','hit_referer','hit_page_path', 'event_category', 'event_label', 'event_value']

df_hits_with_target_action.drop(cols_to_drop, axis=1, inplace=True)
df_hits_with_target_action.head(5)
```

```python
# Merge datasets
df_sessions_with_target = ga_sessions.merge(df_hits_with_target_action, how='outer', on='session_id')
df_sessions_with_target
```

```python
#checking duplicates
df_sessions_with_target.duplicated().sum()
```

```python
#drop duplicates
df_sessions_with_target.drop_duplicates(inplace=True)
```

```python
df_sessions_with_target.shape
```

```python
#save dataset with target action to csv file
df_sessions_with_target.to_csv('../data/df_with_target.csv', index=False)
```
